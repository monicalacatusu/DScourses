## Plot COVID data on a few countries

```python
import pandas as pd
import seaborn as sns

 #provided link#url = 'https://www.ecdc.europa.eu/sites/default/files/documents/COVID-19-geographic-disbtribution-worldwide-2020-03-26.xlsx'
url = 'https://www.ecdc.europa.eu/sites/default/files/documents/COVID-19-geographic-disbtribution-worldwide-2020-07-01.xlsx'

df_raw = pd.read_excel(url)

print(df_raw['countriesAndTerritories'].unique())
df_raw.head()
```

```python
df = df_raw

#df = df[df['Countries and territories']=='Denmark']

df = df[df['countriesAndTerritories'].isin([ 'China','Denmark','Italy','Romania','Spain','United_States_of_America'])]

df = df.sort_values(['year', 'month' ,'day'], ascending=[1, 1, 1])

df['ts'] = pd.to_datetime(df[['year', 'month' ,'day']])
```

```python
df = df.rename(columns = {'countriesAndTerritories':'Countries and territories','cases':'Cases','deaths':'Deaths'})
```

```python
sns.lineplot(x='ts', y='Cases', data=df, hue='Countries and territories')
```

```python
sns.lineplot(x='ts', y='Deaths', data=df, hue='Countries and territories')
# Something is shady in Spain (data)
```
